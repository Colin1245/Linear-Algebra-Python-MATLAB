```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
from mpl_toolkits.mplot3d import Axes3D
```

## 140.eigenvalues of diagonal and triangular matrices

```python
A = np.diag([234, 1234])
print(A), print(' ')
print(np.linalg.eig(A)[0])
print(' ')
```

```python
# expand to NxN diagonal
A = np.diag(np.arange(3,18))
print(A),print(' ')
print(np.linalg.eig(A)[0])
print(' ')
```

```python
# repeat with triangular (lower/upper)
A = np.triu( np.random.randn(4,4) )

print(A),print(' ')
print(np.linalg.eig(A)[0])
```

## 141.eigenvalues of random matrices

```python
for i in range(200):
    A = np.random.randn(40,40)
    evals = np.linalg.eig(A)[0]
    plt.plot(np.real(evals),np.imag(evals),'s',markersize = 1)
    
plt.axis('square')
plt.show()
```
